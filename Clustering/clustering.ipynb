```python
import sklearn
from numpy import where
from sklearn.datasets import make_classification
from matplotlib import pyplot
from numpy import unique
```

```python
X, y=make_classification(
    n_samples=1000,        # number of samples
    n_features=2,          # number of features
    n_informative=2,        # number of informative features
    n_redundant=0,          # number of redundant features
    n_clusters_per_class=1, # number of clusters per class
    random_state=4,         # 
)

# create scatter plot for samples from each class
for class_value in range(2):
    # get row indexes for samples with this class
    row_ix = where(y == class_value)
    # create scatter of these samples
    pyplot.scatter(X[row_ix,0],X[row_ix,1])
# show the plot
pyplot.show()
```

```python
"""
Affiity Propagation

Affinity Propagation involces finding a set of exemplars that best summarize the data

"""
from sklearn.cluster import AffinityPropagation

# define the model
model = AffinityPropagation(damping=0.9, random_state=None)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix,0],X[row_ix,1])
pyplot.show()
```

```python
"""
Agglomerative clustering

Agglomerative clustering involves merging examples until the desired numver os clusters is achieved.
It is a part of a broader class of hierarchical clusterin methods and you can learn more here:
https://en.wikipedia.org/wiki/Hierarchical_clustering
"""
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=2)

yhat = model.fit_predict(X)
clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

```python
"""
BIRCH Clustering (Balanced Iterative Reducing and Clustering using)

BIRCH involves constructing a tree structure from hich cluster centroids are extracted.
"""
from sklearn.cluster import Birch

model = Birch(threshold=0.01, n_clusters=2)
model.fit(X)
yhat = model.predict(X)
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat==cluster)
    pyplot.scatter(X[row_ix, 0],X[row_ix, 1])
pyplot.show()
```

```python
"""
DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

DBSCAN involves finding high-density area in the domain and expanding those area of the feature space around them as clusters.
"""
from sklearn.cluster import DBSCAN

model = DBSCAN(eps = 0.30, min_samples=9)
yhat = model.fit_predict(X)
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat==cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

```python
"""
K-Means

K-Means may be the most widely know clustering algorithm and involver assigning examples to clusters in an effor to minimize the variance within each cluster
"""
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)
model.fit(X)
yhat = model.predict(X)
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat==cluster)
    pyplot.scatter(X[row_ix,0],X[row_ix,1])
pyplot.show()
```

```python
"""
Mini-Batch K-Means

Mini-Batch K-Means is a modified of k-means that makes update to the cluster centroids using mini-batches of samples rather than the entire dataset, which can make it faster for large datasets and perhaps more robust to statistical moise.
"""
from sklearn.cluster import MiniBatchKMeans

model = MiniBatchKMeans(n_clusters=2, batch_size = 4096)
model.fit(X)
yhat = model.predict(X)
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat==cluster)
    pyplot.scatter(X[row_ix,0],X[row_ix,1])
pyplot.show()
```

```python
"""
Mean Shift

Mean Shift clustering involves findin and adapting centroids based on the density of examples in the feature space.
"""

from sklearn.cluster import MeanShift

model = MeanShift()
yhat = model.fit_predict(X)
clusters = unique(yhat)

for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix,0],X[row_ix,1])
pyplot.show()
```

```python
"""
OPTICS (Ordering Points To Identify the Clusterind Structure)
 
OPTICS clustering is a modified version of DBSCAN described above
"""
from sklearn.cluster import OPTICS

model = OPTICS(eps=0.8, min_samples=10)
yhat = model.fit_predict(X)
clusters = unique(yhat)
for cluster in clusters:
	row_ix = where(yhat == cluster)
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

```python
"""
Spectral Clustering

Spectral Clustering is a general class of clustering methods, drawn from linear algebra
"""
from sklearn.cluster import SpectralClustering

model = SpectralClustering(n_clusters=2)
yhat = model.fit_predict(X)
clusters = unique(yhat)

for cluster in clusters:
	row_ix = where(yhat == cluster)
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

```python
"""
Guassian Mixture Model

A Guassian Mixture model summarizes a multivariate probability density function with a mixture of Gaussian probability distributions as its name suggests. 
"""
from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components=2)
model.fit(X)
yhat = model.predict(X)
clusters = unique(yhat)

for cluster in clusters:
	row_ix = where(yhat == cluster)
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```
